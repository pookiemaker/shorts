```python
import urllib3 
import json
def api_fundamentals(symbol, api_key):
    # Fundamentals API
    url = 'www.alphavantage.co'
    querystring = '/query?function=OVERVIEW&symbol={0}&apikey={1}'.format(symbol, api_key)
    http = urllib3.PoolManager()
    r = http.request('GET', 'https://'+url+querystring)
    data = json.loads(r.data.decode('utf-8'))
    df = pd.DataFrame.from_records(data, index=[symbol])
    return df
api_key='xxxyyyzzz'
df = api_fundamentals('GME',api_key)
```
