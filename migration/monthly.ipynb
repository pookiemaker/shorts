```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd

#sdir =  '/mnt/5c823cd2-502b-4990-a220-579838be1f10/data/monthyregsho/'
sdir =  '/mnt/5c823cd2-502b-4990-a220-579838be1f10/data/test/'
filename = 'gme2.txt'
```

```python
f = open(sdir+filename)
df = pd.read_csv(f,sep="|")
```

```python
 #df.sort_values(by=df['Time'])
```

```python
df['Date'] = pd.to_datetime(df['Date'],format='%Y%m%d')
df['Time'] = pd.to_datetime(df['Time']) 
```

```python
plt.figure(figsize=(15,5))

plt.plot(df['Time'],df['Price'],'b.-')
plt.title('Short Sales on 2021-07-27')



plt.figure(figsize=(15,5))

plt.plot(df['Time'],df['Size'],'r')
plt.xticks(rotation=45)


plt.title('Short Sales on 2021-07-27')

```
